### Question 1
What percent of majors are predominatly male? Predomindately female?

***Hypothesis***: A higer percent of majors are predominantly female. 

The distribution of majors in men seems more skewed to the left than female distrition of majors in the histograms above. Suggesting that males may dominate certain majors by a heavy margin, but females are more evenly spread out among majors than men.


```python
subset= recent_grads[['Major','ShareWomen','Men','Women']]
subset_male = subset[subset['ShareWomen'] < .50]
subset_male = subset_male.sort_values(by='ShareWomen', ascending=True)
subset_male.tail()
```

```python
subset_male.shape
```

```python
subset_female = subset[subset['ShareWomen'] > .50]
subset_female = subset_female.sort_values(by='ShareWomen', ascending=False)
subset_female.tail()
```

```python
total_female = subset_female.shape
total_female = float(total_female[0])
total_female
```

```python
total_male = subset_male.shape
total_male = float(total_male[0])
total_male
```

```python
total = float(subset_male.shape[0]) + float(subset_female.shape[0])
total
```

```python
percent_male = (total_male / total) * 100
percent_male
```

```python
percent_female = (total_female / total) * 100
percent_female
```

### Results for Question 1
The data analysis method above reveals that a higher percentage of majors are predominatley female. Females dominate in 55% of majors analyzed. Males dominate 44% in the majors analyzed. Roughly a 11% difference between females and males.
